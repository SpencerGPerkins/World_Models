```python
import torch
from torch.utils.data import Dataset
import numpy as np


class ReachDataset(Dataset):

    def __init__(self, path):

        data = np.load(path)

        self.images = data["images"]
        self.states = data["robot_states"]
        self.actions = data["actions"]
        self.next_images = data["next_images"]
        self.next_states = data["next_robot_states"]


    def __len__(self):
        return len(self.actions)


    def __getitem__(self, idx):

        return {
            "image": torch.tensor(
                self.images[idx]
            ).permute(2,0,1).float()/255,

            "state": torch.tensor(
                self.states[idx]
            ).float(),

            "action": torch.tensor(
                self.actions[idx]
            ).float(),

            "next_image": torch.tensor(
                self.next_images[idx]
            ).permute(2,0,1).float()/255,

            "next_state": torch.tensor(
                self.next_states[idx]
            ).float()
        }
```
